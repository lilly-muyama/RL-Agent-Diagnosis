```python
import argparse
import multiprocessing
import pandas as pd
import numpy as np
```

```python
def run_script_with_seed(seed):
    parser = argparse.ArgumentParser()
    parser.add_argument('--seed', type=int, help='Random seed value')
    
    # Convert the seed to a string argument
    args = parser.parse_args(['--seed', str(seed)])
    
    # Run the script with the specified seed
    %run -i run_script.py --seed $args.seed
```

```python
seeds = [42, 63, 84, 105, 126]
```

```python
if __name__ == '__main__':
    with multiprocessing.Pool(processes=len(seeds)) as pool:
        pool.map(run_script_with_seed, seeds)
```
