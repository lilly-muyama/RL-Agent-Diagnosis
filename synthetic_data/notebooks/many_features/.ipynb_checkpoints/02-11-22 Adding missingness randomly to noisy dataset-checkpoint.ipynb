```python
def insert_nans(df, col, frac):
    non_null_indices = df[df[col].notna()].index.tolist() #getting not null indices
    to_change_indices = random.sample(non_null_indices, np.floor(frac*len(non_null_indices)))
    df = df.loc[to_change_indices, col]
    return df
```

```python
for column in df.columns:
    if column not in ['hemoglobin', 'gender', 'label']:
        df[column] = df[column].sample(frac=1-frac)
df.head()
```
